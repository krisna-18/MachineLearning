```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
# Mendefinisikan data baru
countries = ("Indonesia", "Malaysia", "Singapore", "Thailand", "Vietnam")
gold_medals = (1, 0, 2, 1, 1)

# Membuat Bar Chart 
plt.bar(x=countries, height=gold_medals)
plt.xticks(rotation=45)
plt.show()
```

```python
# Mengurutkan data berdasarkan jumlah medali emas
df = pd.DataFrame({
    "Countries" : countries,
    "Gold Medals" : gold_medals,
})

df.sort_values(by="Gold Medals", inplace=True)

plt.barh(y=df["Countries"], width=df["Gold Medals"])
plt.xlabel("Gold Medals")
plt.title("Gold Medals of Countries in Southeast Asia in Summer Olympics 2020")
plt.show()
```

```python
# Membuat BarChart menggunakan seaborn
sns.barplot(y=df["Countries"], x=df["Gold Medals"], orient="h", color="gold")
plt.xlabel("Gold Medals")
plt.title("Gold Medals of Countries in Southeast Asia in Summer Olympics 2020")
plt.show()
```
