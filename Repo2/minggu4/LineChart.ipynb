```python
import matplotlib.pyplot as plt
import pandas as pd

# Mendefinisikan data baru
books = ("Harry Potter", "The Hobbit", "To Kill a Mockingbird", "1984", "The Great Gatsby")
sales = (1200, 1500, 800, 2000, 1700)

# Membuat Line Chart 
plt.plot(books, sales, marker='o')
plt.xticks(rotation=45)
plt.title('Book Sales in a Month')
plt.xlabel('Books')
plt.ylabel('Sales')
plt.show()
```

```python
# Mengubah data menjadi DataFrame
df = pd.DataFrame({
    "Books" : books,
    "Sales" : sales,
})
```

```python
# Mengurutkan data berdasarkan penjualan
df.sort_values(by="Sales", inplace=True)
```

```python
# Membuat Line Chart dari data yang telah diurutkan
plt.plot(df["Books"], df["Sales"], marker='o', color='green')
plt.xticks(rotation=45)
plt.title('Book Sales in a Month (Sorted)')
plt.xlabel('Books')
plt.ylabel('Sales')
plt.show()
```
